```python
#We need to get rid of the \t and | chars present in the current train data file
data_train = []
split_chars = ['\t', '|']
with open('/kaggle/input/music-recommender-2023-spring/trainItem2.txt', 'r') as file:
    for line in file:
        line = line.strip()
        line_data = []
        for char in split_chars:
            if char in line:
                line_data = line.split(char)
                break
        else: 
            line_data = [line]
        data_train.append(line_data)
data_train[0:10]
```

```python
#We now need to convert from string to integer
list_train = []
for line in data_train: 
    val = [int(x) for x in line]
    list_train.append(val)
list_train[0:40]
```

```python
userID_temp = list_train[0][0]
userID_dataPoints = list_train[0][1]

iteration = 0
for row in range(len(list_train)):
    if iteration == 0:
        iteration += 1
        continue
    elif(iteration > 0 and iteration <= userID_dataPoints):
        list_train[row].insert(0, userID_temp)
        iteration += 1
    else:
        iteration = 1
        userID_temp = list_train[row][0]
        userID_dataPoints = list_train[row][1]
        
#This matrix comprises of all user ratings in our database
user_train_matrix = list_train
user_train_matrix[0:37]
```
